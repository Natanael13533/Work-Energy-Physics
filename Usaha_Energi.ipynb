```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

Soal Fisika

Dian pulls a backpack on wheels down 100 m hall. The 60 N force is applied at an angle of 60 degree above the horizontal. How much work is done by dian?

```python
# Diketahui
gaya = 60  # Gaya
jarak = 100  # Jarak 
sudut_int = 60  # Masih belum dalam bentuk sudut
```

```python
print(type(sudut_int))
```

W=F⋅d⋅cos(θ)

- W adalah usaha (joules)
- F adalah gaya yang dihasilkan (N)
- d adalah jarak perpindahan (m)
- θ adalah sudut antara gaya dan perpindahan

```python
# ubah sudut dari int menjadi radian
sudut = math.radians(sudut_int)
```

```python
print(type(sudut))
```

```python
usaha = gaya * jarak * math.cos(sudut)
usaha
```

Perbandingan graph hasil dari usaha dengan gaya, jarak, sudut yang berbeda

masing usaha akan dihitung dengan cara :

- usaha dihitung dengan sudut berbeda, namun gaya dan juga jarak sama dengan soal yang sudah dikerjakan
- usaha dihitung dengan jarak berbeda, namun gaya dan sudut sama dengan soal yang sudah dikerjakan
- usaha dihitung dengan gaya berbeda, namun jarak dan sudut sama dengan soal yang sudah dikerjakan

masing-masing hasil nantinya akan di tampilkan menggunakan graph

```python
# data untuk graph
angles = np.linspace(0, 90, 100)  # sudut dari 0 sampau 90 derajat
forces = np.linspace(10, 100, 100)  # gaya dari 10 N sampai 100 N
distances = np.linspace(10, 200, 100)  # jarak dari 10 m sampai 200 m
```

```python
# ubah angles ke radians
angles_radians = np.radians(angles)
```

```python
# Hitung usaha untuk masing-masing jarak
usaha_angles = gaya * jarak * np.cos(angles_radians)
usaha_angles
```

```python
# Hitung usaha untuk masing-masing gaya
usaha_forces = forces * jarak * np.cos(sudut)
usaha_forces
```

```python
# Calculate work done for varying distances
usaha_distances = gaya * distances * np.cos(sudut)
usaha_distances
```

Setelah menghitung usaha, saya akan menampilkan sebuah graph.

```python
# Plotting
plt.figure(figsize=(18, 6))
```

```python
# usaha vs. Sudut
plt.subplot(1, 2, 2)
plt.plot(angles, usaha_angles, label='Work done', color='blue')
plt.xlabel('Angle (degrees)')
plt.ylabel('Work Done (Joules)')
plt.title('Work Done vs. Angle')
plt.grid(True)
plt.legend()
```

```python
# Usaha vs. Gaya
plt.subplot(1, 2, 2)
plt.plot(forces, usaha_forces, label='Work done', color='green')
plt.xlabel('Force (N)')
plt.ylabel('Work Done (Joules)')
plt.title('Work Done vs. Force')
plt.grid(True)
plt.legend()
```

```python
# Work done vs. Distance
plt.subplot(1, 2, 2)
plt.plot(distances, usaha_distances, label='Work done', color='red')
plt.xlabel('Distance (m)')
plt.ylabel('Work Done (Joules)')
plt.title('Work Done vs. Distance')
plt.grid(True)
plt.legend()
```

dari hasil graph dapat dilihat hasil dari usaha dengan masing gaya, jarak, dan sudut yang berbeda.

- untuk sudut dari 0 sampai 90 : bisa dilihat bahwa hasil usaha semakin naik jika sudut semakin besar
- untuk gaya dari 10 sampai 100 : hasil usaha juga akan semakin naik
- untuk jarak dari 0 sampai 200 : hasil usaha juga semakin naik sama seperti gaya
